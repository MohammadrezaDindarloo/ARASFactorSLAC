```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import csv
import os
direc = os.getcwd()
```

# Result

```python
cutoff = 30
```

### result for uwb

```python
rows = []
with open(direc + "/Dataset150_v1/simuation_uwb_result.csv", 'r') as file:
    csvreader = csv.reader(file)
    header = next(csvreader)
    rows.append(header)
    for row in csvreader:
        if float(row[4]) < cutoff:
            rows.append(row)
error_pulley_sim_uwb = [float(row[0]) for row in rows]
error_pulley_sim_uwb = [float(row[1]) for row in rows]
error_pulley_sim_uwb = [float(row[2]) for row in rows]
error_pulley_sim_uwb = [float(row[3]) for row in rows]
error_pulley_sim_sum_uwb = [float(row[4]) for row in rows]
plt.hist(error_pulley_sim_sum_uwb)
plt.show()
```
