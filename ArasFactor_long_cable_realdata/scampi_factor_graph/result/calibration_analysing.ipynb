```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import csv
import os
import seaborn as sns

direc = os.getcwd()
```

### result for uwb

```python
cutoff = 100
rows = []
with open(direc + "/calibration_result.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        if float(row[4]) < cutoff:
            rows.append(row)
error_pulley_a = [float(row[0]) for row in rows]
error_pulley_b = [float(row[1]) for row in rows]
error_pulley_c = [float(row[2]) for row in rows]
error_pulley_d = [float(row[3]) for row in rows]
error_pulley_sum = [float(row[4]) for row in rows]

# Plot histograms with shading
plt.hist(error_pulley_sum, alpha=0.5, label='Pulley Sum')
plt.hist(error_pulley_a, alpha=0.5, label='Pulley A')
plt.hist(error_pulley_b, alpha=0.5, label='Pulley B')
plt.hist(error_pulley_c, alpha=0.5, label='Pulley C')
plt.hist(error_pulley_d, alpha=0.5, label='Pulley D')

plt.legend()  # Add legend
plt.show()
```

```python
print(np.mean(error_pulley_sum))
print(np.mean(error_pulley_a))
print(np.mean(error_pulley_b))
print(np.mean(error_pulley_c))
print(np.mean(error_pulley_d))
print(np.mean(error_pulley_a)+np.mean(error_pulley_b)+np.mean(error_pulley_c)+np.mean(error_pulley_d))
```

```python
rows = []
with open(direc + "/calibration_result.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        if float(row[4]) < cutoff:
            rows.append(row)
error_pulley_a = [float(row[0]) for row in rows]
error_pulley_b = [float(row[1]) for row in rows]
error_pulley_c = [float(row[2]) for row in rows]
error_pulley_d = [float(row[3]) for row in rows]
error_pulley_sum = [float(row[4]) for row in rows]

# Set up the plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.hist(error_pulley_sum, bins=20, alpha=0.7, color='blue', edgecolor='black', label='Pulley Sum')
plt.hist(error_pulley_a, bins=20, alpha=0.5, color='green', edgecolor='black', label='Pulley A')
plt.hist(error_pulley_b, bins=20, alpha=0.5, color='orange', edgecolor='black', label='Pulley B')
plt.hist(error_pulley_c, bins=20, alpha=0.5, color='red', edgecolor='black', label='Pulley C')
plt.hist(error_pulley_d, bins=20, alpha=0.5, color='purple', edgecolor='black', label='Pulley D')

# Add labels and title
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.title('Histogram of Error for Different Pulleys')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Save or show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.savefig('histogram.png', dpi=300)  # Save plot as PNG with higher resolution for the paper
plt.show()
```
