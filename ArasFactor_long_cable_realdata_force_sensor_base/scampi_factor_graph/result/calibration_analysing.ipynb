```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def plot_3d_points(original_points, Optimized_points, color='blue'):
  """
  Plots points in 3D space with their specified neighbors.

  Args:
      original_points: A list of lists, where each inner list represents a point (x, y, z).
      Optimized_points: A list of lists, where each inner list represents a neighbor point (x, y, z).
      color: The color for the points (default: 'blue').
  """

  fig = plt.figure(figsize=(8, 6))
  ax = fig.add_subplot(111, projection='3d')

  # Plot original points
  x_original, y_original, z_original = zip(*original_points)
  ax.scatter(x_original, y_original, z_original, c=color, marker='o',alpha=1, label='Original Points')

  # Plot neighbor points
  x_neighbor, y_neighbor, z_neighbor = zip(*Optimized_points)
  ax.scatter(x_neighbor, y_neighbor, z_neighbor, c='red', marker='s',alpha=1, label='Neighbor Points')

  # Set labels and title
  ax.set_xlabel('X(m)')
  ax.set_ylabel('Y(m)')
  ax.set_zlabel('Z(m)')
  ax.set_title('3D Points with Neighbors')
  ax.legend()

  plt.show()

# Example usage
pulley_attach_point_1 = [-125.0, -110.0, 48.0]
pulley_attach_point_2 = [125.0, -110.0, 48.0]
pulley_attach_point_3 = [125.0,  110.0, 48.0]
pulley_attach_point_4 = [-125.0,  110.0, 48.0]

original_points = [pulley_attach_point_1, pulley_attach_point_2, pulley_attach_point_3, pulley_attach_point_4]
Optimized_points = [[-124.824, -109.847,  47.9208], [124.859, -109.867,  47.9092], [125.079,   110.06,  47.9763], [-124.956,  109.966,  47.9474]]  # Specific neighbor coordinates

plot_3d_points(original_points, Optimized_points)
```
