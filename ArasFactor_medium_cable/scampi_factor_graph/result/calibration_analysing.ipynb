```python

# Set seaborn style for improved aesthetics
sns.set_style("whitegrid")

# Initialize list to store data rows
rows = []

# Read data from CSV file
with open(direc + "/calibration_result_500_Final.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        # Filter rows based on cutoff value
        if float(row[4]) < cutoff:
            rows.append(row)

# Extract error data for each pulley
error_pulley_a = [float(row[0]) for row in rows]
error_pulley_b = [float(row[1]) for row in rows]
error_pulley_c = [float(row[2]) for row in rows]
error_pulley_d = [float(row[3]) for row in rows]
error_pulley_sum = [float(row[4]) for row in rows]

# Set up the plot
plt.figure(figsize=(10, 8))  # Adjust figure size as needed

# Plot histograms for each type of pulley error with enhanced visual features
plt.hist(error_pulley_sum, bins=20, alpha=0.7, color='skyblue', edgecolor='black', label='Sum of pulley error')
plt.hist(error_pulley_a, bins=20, alpha=0.5, color='lightgreen', edgecolor='black', label='Pulley 1 error')
plt.hist(error_pulley_b, bins=20, alpha=0.5, color='coral', edgecolor='black', label='Pulley 2 error')
plt.hist(error_pulley_c, bins=20, alpha=0.5, color='salmon', edgecolor='black', label='Pulley 3 error')
plt.hist(error_pulley_d, bins=20, alpha=0.5, color='plum', edgecolor='black', label='Pulley 4 error')

# Add labels and title
plt.xlabel('Error (mm)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
# plt.title('Histogram of Error for Different Pulleys', fontsize=16)

# Add legend with improved fontsize
plt.legend(fontsize=12)

# Save or show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.savefig('histogram.png', dpi=300)  # Save plot as PNG with higher resolution for the paper
plt.show()
```
