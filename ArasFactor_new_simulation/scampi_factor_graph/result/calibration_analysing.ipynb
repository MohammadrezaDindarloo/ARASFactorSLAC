```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load data from the CSV file
cutoff = 50
rows = []
with open(direc + "/calibration_result_500.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        if float(row[4]) < cutoff:
            rows.append(row)
error_pulley_a = [float(row[0]) for row in rows]
error_pulley_b = [float(row[1]) for row in rows]
error_pulley_c = [float(row[2]) for row in rows]
error_pulley_d = [float(row[3]) for row in rows]
error_pulley_sum = [float(row[4]) for row in rows]

# Set up the plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

# Plot histograms
plt.hist(error_pulley_sum, bins=20, color='skyblue', alpha=0.7, label='Pulley Sum', density=True)
plt.hist(error_pulley_a, bins=20, color='salmon', alpha=0.7, label='Pulley A', density=True)
plt.hist(error_pulley_b, bins=20, color='lightgreen', alpha=0.7, label='Pulley B', density=True)
plt.hist(error_pulley_c, bins=20, color='orchid', alpha=0.7, label='Pulley C', density=True)
plt.hist(error_pulley_d, bins=20, color='gold', alpha=0.7, label='Pulley D', density=True)

# Plot Gaussian bell-like lines
x_range = np.linspace(-10, 10, 1000)
plt.plot(x_range, norm.pdf(x_range, np.mean(error_pulley_sum), np.std(error_pulley_sum)), color='skyblue', linestyle='--')
plt.plot(x_range, norm.pdf(x_range, np.mean(error_pulley_a), np.std(error_pulley_a)), color='salmon', linestyle='--')
plt.plot(x_range, norm.pdf(x_range, np.mean(error_pulley_b), np.std(error_pulley_b)), color='lightgreen', linestyle='--')
plt.plot(x_range, norm.pdf(x_range, np.mean(error_pulley_c), np.std(error_pulley_c)), color='orchid', linestyle='--')
plt.plot(x_range, norm.pdf(x_range, np.mean(error_pulley_d), np.std(error_pulley_d)), color='gold', linestyle='--')

# Add labels and title
plt.xlabel('Error', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.title('Histogram of Error for Different Pulleys with Gaussian Bell-Like Lines', fontsize=16)

# Add legend
plt.legend(fontsize=12)

# Set x-axis limit to the right side only
plt.xlim(0, 10)

# Save or show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.savefig('histogram_with_gaussian.png', dpi=300)  # Save plot as PNG with higher resolution for the paper
plt.show()
```
